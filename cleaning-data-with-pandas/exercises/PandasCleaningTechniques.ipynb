## Irregular Data

With missing data out of the way, turn your attention to any outliers. Just as we did for missing data, we first need to detect the outliers.

```python
# Keep an eye out for outliers!
```

```python
# Generates descriptive statistics of the Age column
data_file['Age'].describe()
```

```python
# Returns the IQR, and the age range to detect outliers
q1 = data_file['Age'].quantile(0.25)
q3 = data_file['Age'].quantile(0.75)
IQR = q3-q1
print(IQR)

lower_age = q1 - 1.5 * IQR
print(lower_age)

upper_age = q3 + 1.5 * IQR
print(upper_age)
```

```python
# Returns the rows in the Age column that are outliers
data_file[(data_file.Age < lower_age) | (data_file.Age > upper_age)]
```

```python
# Returns rows in the dataframe that do not have outliers in the Age column
new_df = data_file[(data_file.Age > lower_age) & (data_file.Age < upper_age)]
new_df
```
